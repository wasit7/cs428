```python
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

$f(x)=sinc(x)$

$f'(x)=\frac{\cos x}{x} - \frac{\sin x}{x^2}$

```python
def f(x):
    return np.sin(x)

def df(x):
    return np.cos(x)

list_atx=np.arange(-5.0,5.0,0.01)
list_aty=f(list_atx)
plt.style.use('ggplot')
plt.plot(list_atx, list_aty,'-b')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('$f(x)=sinc(x)$')
```

```python
x=0.7
list_x=[x]
list_y=[f(x)]
for i in range(10):
    x = x - f(x)/df(x)
    list_x.append(x)
    list_y.append(f(x))
    print("x:{}, f(x):{}".format(x,f(x)))
    
list_atx=np.arange(0,10,0.01)
list_aty=f(list_atx)
plt.style.use('ggplot')
plt.plot(list_atx, list_aty,'-b')
plt.plot(list_x, list_y,'-ro')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('$f(x)=sinc(x)$')
```

```python
x=0.8
list_x=[x]
list_y=[f(x)]
e=0.1
for i in range(100):
    x = x - e*df(x)
    list_x.append(x)
    list_y.append(f(x))
    print("x:{}, f(x):{}".format(x,f(x)))
    
list_atx=np.arange(0,10,0.01)
list_aty=f(list_atx)
plt.style.use('ggplot')
plt.plot(list_atx, list_aty,'-b')
plt.plot(list_x, list_y,'-ro')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('$f(x)=sinc(x)$')
```
