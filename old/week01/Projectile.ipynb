# Projectile simulation


* $s_x,s_y$ position
* $u_x,u_y$velocity
* $a_x,a_y$accelelation

```python
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
#init
sx=0
sy=0
ux=4
uy=3
ax=0
ay=-9.8
t=0.0
dt=0.1
g=-9.8
posx=[0]
posy=[0]
while sy>=0:
    t+=dt #t=t+dt
    ux=ux
    sx=sx+ux*dt
    uy=uy+g*dt
    sy=sy+uy*dt
    posx.append(sx)
    posy.append(sy)
print("ux:{} uy:{} sx:{} sy:{}".format(ux,uy,sx,sy))
```

```python
plt.plot(posx,posy,'--ro')
```
