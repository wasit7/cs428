# Sigmiod (Logistic function)
$f(x)=\frac{e^x}{e^x+1}$

```python
def sigmoid(x):
    y=np.exp(x)/(np.exp(x)+1)
    return y
```

```python
z=sigmoid(x)
```

```python
plt.style.use('ggplot')
plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('tanh(x) and sigmoid(x)')
```

```python
x=np.arange(-np.pi,np.pi, 0.1)
y=np.sin(x)
plt.plot(x,y,'o')
```
