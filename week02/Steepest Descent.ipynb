# Finding turning points of a function

$f(x)=x^3-2x^2+2$

Turning point $x=0, 4\frac{1}{3}$

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
def f(x):
    return np.power(x,3)-2.0*np.power(x,2) + 2.0

def df(x):
    return 3.0*np.power(x,2)-4.0*x
```

```python
list_atx=np.arange(-1.0,2.5,0.001)
list_aty=f(list_atx)
plt.style.use('ggplot')
plt.plot(list_atx, list_aty,'-b')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('$f(x)=x^3-2x^2+2$')
```

Update $x$ by steepest descent

$x_i = x_{i-1} - \epsilon.{f}'(x_{i-1})$

```python
x=2.0
list_x=[x]
list_y=[f(x)]
epsilon = 0.1
for i in range(10):
    x = x - epsilon*df(x)
    list_x.append(x)
    list_y.append(f(x))
    print("x:{}, f(x):{}".format(x,f(x)))
```

```python
list_atx=np.arange(1.25,2.25,0.001)
list_aty=f(list_atx)
plt.style.use('ggplot')
plt.plot(list_atx, list_aty,'-b')
plt.plot(list_x, list_y,'-ro')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Steepest Descent of $f(x)=x^3-2x^2+2$')
```

```python
x=0.5
list_x=[x]
list_y=[f(x)]
epsilon = 0.1
for i in range(10):
    x = x + epsilon*df(x)
    list_x.append(x)
    list_y.append(f(x))
    print("x:{}, f(x):{}".format(x,f(x)))
```

```python
list_atx=np.arange(-0.1,1.4,0.001)
list_aty=f(list_atx)
plt.style.use('ggplot')
plt.plot(list_atx, list_aty,'-b')
plt.plot(list_x, list_y,'-ro')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Steepest Descent of $f(x)=x^3-2x^2+2$')
```
