```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

def f(x):
    return 2*np.power(x,3)-4.0*np.power(x,2) + 1.0


list_atx=np.arange(-1.0,2.5,0.001)
list_aty=f(list_atx)
plt.style.use('ggplot')
plt.plot(list_atx, list_aty,'-b')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('$f(x)=2x^3-4x^2+1$')
```
